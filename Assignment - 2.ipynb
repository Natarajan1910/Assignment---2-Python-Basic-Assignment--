```python
11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

They are all do the same task. The range(10) call ranges from 0 up to (but not including) 10, range(0, 10) explicitly tells the loop to start at 0, and range(0, 10, 1) explicitly tells the loop to increase the variable by 1 on each iteration.
```
